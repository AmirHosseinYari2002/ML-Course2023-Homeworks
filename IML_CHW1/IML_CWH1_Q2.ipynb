## Plot
>You must plot the cos(x) curve. plot 20 cycle of the curve. the frequency of first half of curve is 50 Hz and frequency of second lalf is 200 Hz. 
> set the x_label, y_label and title for your plot.

```python
### START CODE HERE ###
# f1 -> 0.2s   f2 -> 0.05s
def f(x):
    y = []
    for i in x:
        if i<0.2:
            y.append(np.cos(2*np.pi*50*i))
        else:
            y.append(np.cos(2*np.pi*200*i))
    return y
t = np.arange(0,0.25,0.0001)
y = f(t)
plt.plot(t,y)
plt.xlabel("time")
plt.ylabel("cos(x)")
plt.title("frequency of first half is 50 and seccond half is 200")
### END CODE HERE ###
```
